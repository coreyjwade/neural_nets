# Sequential Deep Learning - Regression, Classification

## Regression

```
# Import libraries and establish DataFrame
# We are using sklearn's housing dataset
import pandas as pd
import numpy as np
from sklearn import datasets
housing = datasets.fetch_california_housing()
df = pd.DataFrame(data= np.c_[housing['data'], housing['target']],columns= housing['feature_names'] + ['target'])
```

```
# Display the first 5 rows
df.head()
```

```
# Display general statistics
df.describe()
```

```
# Check for null values and numeric columns
df.info()
```

```
# Split the Data into Predictor Columns, and Target columns
X = df.iloc[:, :-1]
y = df.iloc[:, -1]
```

```
# Import train_test_split
from sklearn.model_selection import train_test_split

# Split the data into a training and test set
X_train, X_test, y_train, y_test = train_test_split(X, y)
```

```
X.shape
```

```
# Import the Sequential model and Dense layer
from keras.models import Sequential
from keras.layers import Dense

num_cols = X.shape[1]

# Create a Sequential model
model = Sequential()

# Add an input layer and a hidden layer with 10 neurons
model.add(Dense(10, input_shape=(num_cols,), activation="relu"))

# Add a 1-neuron output layer
model.add(Dense(1))

# Summarise your model
print(model.summary())

model.compile(optimizer='adam', loss='mse')

model.fit(X_train, y_train, epochs=30)

# Get score for predictions
score = model.evaluate(X_test, y_test)
    
# Get root mean squared error
rmse = score**0.5
    
# Return root mean squared error
print(rmse)
```

# Classification

```
# 1. Load a DataFrame using the following url and show the first 5 rows
url = 'https://raw.githubusercontent.com/PacktPublishing/Hands-On-Gradient-Boosting-with-XGBoost-and-Scikit-learn-/master/Chapter_2_Decision_Trees_In_Depth/heart_disease.csv'
df = pd.read_csv(url)
df.head()
```

```
# Show distribution of target columnn
df['target'].value_counts()
```

```
# Check for null values and non-numeric columns
df.info()
```

```
# Split data into predctive and target columns
X = df.iloc[:, :-1]
y = df.iloc[:, -1]
```

```
# Split data into training and test set
X_train, X_test, y_train, y_test = train_test_split(X, y)
```

```
# Import the Sequential model and Dense layer
from keras.models import Sequential
from keras.layers import Dense, Dropout

# Prepare number of columns as input to neural network
num_cols = X_train.shape[1]

# Initialize a Sequential model
model = Sequential()

# Add a hidden layer with 8 nodes that takes the number of columns as input
model.add(Dense(8, input_shape=(num_cols,), activation="relu"))
# Add an additional hidden layer with 8 nodes
model.add(Dense(8, activation="relu"))

# Add a 1-neuron output layer
model.add(Dense(1, activation='sigmoid'))

# Summarize your model
print(model.summary())

# Compile the model with an optimizer, loss function, and scoring metric
model.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])

# Fit the model on the training data with epochs (number of rounds) and batch_size (chunking) adjustable
model.fit(X_train, y_train, epochs=100, batch_size=10)

# Check model on the test set
print(model.evaluate(X_test, y_test))
```

```
# Import the Sequential model and Dense layer
from keras.models import Sequential
from keras.layers import Dense, Dropout
from keras.constraints import maxnorm
from keras.callbacks import EarlyStopping

num_cols = X_train.shape[1]

# Create a Sequential model
model = Sequential()

# Add an input layer and a hidden layer with 10 neurons
model.add(Dense(8, input_shape=(num_cols,), activation="relu", kernel_constraint=maxnorm(3)))
model.add(Dropout(0.25))
model.add(Dense(8, activation="relu", kernel_constraint=maxnorm(3)))
model.add(Dropout(0.25))

# Add a 1-neuron output layer
model.add(Dense(1, activation='sigmoid'))

# Summarise your model
print(model.summary())

model.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])

# Define early_stopping_monitor
early_stopping_monitor = EarlyStopping(patience=100)

model.fit(X_train, y_train, epochs=1000, batch_size=20, validation_split=0.2, callbacks=[early_stopping_monitor])

print(model.evaluate(X_test, y_test))
```
